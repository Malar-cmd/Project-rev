```python
df['personalization_level']=df['personalization_level'].fillna('Unknown')
# df['avg_order_value']=df['avg_order_value'].fillna(0.0)
# df['email_open_rate']=df['email_open_rate'].fillna(0.0)
# df['click_through_rate']=df['click_through_rate'].fillna(0.0)
col_names=['avg_order_value','email_open_rate','click_through_rate']
df[col_names]=df[col_names].fillna(0.0)
```

Using box plots and histogram to understand the data distribution

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
```

```python

for col in num_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
for col in num_cols:
    plt.figure(figsize=(6,3))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```
